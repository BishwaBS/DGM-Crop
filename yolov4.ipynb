```python
# define helper functions
def imShow(path):
  import cv2
  import matplotlib.pyplot as plt
  %matplotlib inline

  image = cv2.imread(path)
  height, width = image.shape[:2]
  resized_image = cv2.resize(image,(3*width, 3*height), interpolation = cv2.INTER_CUBIC)

  fig = plt.gcf()
  fig.set_size_inches(18, 10)
  plt.axis("off")
  plt.imshow(cv2.cvtColor(resized_image, cv2.COLOR_BGR2RGB))
  plt.show()
```

```python
##Replace strings
import glob
import fileinput
for file in glob.glob("/content/gdrive/MyDrive/Chapter3/yolov4/darknet/data/Uni_model/val/*.txt"):
  print(file)
  with fileinput.FileInput(file, inplace=True ) as file:
    for line in file:
  
        print(line.replace("2", "0"), end='')
        
    file.close ()
  break
```

```python
#Copy strings from another file, edit it, and write to new file
import glob
import os
import fileinput
for file in glob.glob("/content/gdrive/MyDrive/Chapter3/yolov4/darknet/data/multi_model/test_soybean_GS3_cloudy_multi/*.txt"):
  print(file)
  with open(file, "r") as inputfile, open(os.path.join("/content/gdrive/MyDrive/Chapter3/yolov4/darknet/data/uni_model/test_soybean_GS3_cloudy_uni", os.path.split(file)[1] ), "w") as output:
  
    for line in inputfile:
        strings=line.split()
        if strings[0]=="1" or strings[0]=="2":
          strings[0]="0"
        # print(line.replace("2", "0"), end='')
        print(strings)
        output.write("{} {} {} {} {} {}".format(strings[0], strings[1], strings[2], strings[3], strings[4], "\n"))
        # output.write("\n")
        # output.writelines(strings)
    output.close ()
    inputfile.close()
```
